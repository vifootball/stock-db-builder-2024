```python
import pandas as pd
```

```python
df1 = pd.read_csv("./stock_analysis_com/etf_list_240605_1.csv", delimiter='\t')
df2 = pd.read_csv("./stock_analysis_com/etf_list_240605_2.csv", delimiter='\t')
df3 = pd.read_csv("./stock_analysis_com/etf_list_240605_3.csv", delimiter='\t')
df4 = pd.read_csv("./stock_analysis_com/etf_list_240605_4.csv", delimiter='\t')
df4
```

```python
df_all = []
df_all.append(df1)
df_all.append(df2)
df_all.append(df3)
df_all.append(df4)
df_all = pd.concat(df_all, ignore_index=True)
df_all.rename(columns={
    'Symbol': 'symbol',
    'Fund Name': 'fund_name',
    'Asset Class': 'asset_class',
    'Assets': 'assets'
}, inplace=True)
df_all
```

```python
df_all.to_csv('./stock_analysis_com/etf_list_240605_union.csv', index=False)
```
